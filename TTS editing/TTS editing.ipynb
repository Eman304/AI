```python
import json
import edge_tts
from IPython.display import Audio, display
import os
```

```python

# ุงุฎุชูุงุฑ ุงููููุงุช
files = [
    r"C:\Users\EMAM ABD EL MONSEF\Desktop\Eman\AI\TTS\stories_ar.json",
    r"C:\Users\EMAM ABD EL MONSEF\Desktop\Eman\AI\TTS\stories_en.json"
]

# ุฅูุดุงุก ูุฌูุฏุงุช ููุญูุธ
os.makedirs("audio_stories", exist_ok=True)
os.makedirs("text_stories", exist_ok=True)

# ุชุญุฏูุฏ ุงููุบุฉ ูู ุงุณู ุงูููู
def detect_language_from_filename(filename):
    if "_ar" in filename:
        return "ar"
    elif "_en" in filename:
        return "en"
    else:
        return "ar"
```

```python
# ูุฑุงุกุฉ ุงููุตุต ูู ุงููููุงุช (ุณูุงุก dict ุฃู list)
stories = {}
for file in files:
    lang = detect_language_from_filename(file)
    with open(file, "r", encoding="utf-8") as f:
        file_stories = json.load(f)

        # ูู dict ูุจุงุดุฑุฉ
        if isinstance(file_stories, dict):
            for title, content in file_stories.items():
                stories[title] = (content, lang)

        # ูู list ูููุง {title, content}
        elif isinstance(file_stories, list):
            for story in file_stories:
                title = story.get("title", "ูุตุฉ ุจุฏูู ุงุณู")
                content = story.get("content", "")
                stories[title] = (content, lang)
```

```python
# ุนุฑุถ ุงูุนูุงููู
print("๐ ุงููุตุต ุงููุชุงุญุฉ:")
for i, title in enumerate(stories.keys(), 1):
    print(f"{i}. {title}")

```

```python
# ุงุฎุชูุงุฑ ุงููุตุฉ
choice = int(input("\nุงูุชุจ ุฑูู ุงููุตุฉ ุงููู ุนุงูุฒ ุชุณูุนูุง: "))
selected_title = list(stories.keys())[choice - 1]
selected_content, lang = stories[selected_title]
```

```python
# ุงุฎุชูุงุฑ ููุน ุงูุตูุช
print("\n๐๏ธ ุงุฎุชุฑ ููุน ุงูุตูุช:")
print("1 - ููุฏ ๐ฆ")
print("2 - ุจูุช ๐ง")
print("3 - ุตูุช ูุฑุชููู ๐ญ")

voice_choice = int(input("ุงูุชุจ ุฑูู ุงุฎุชูุงุฑู: "))
```

```python
# ุฃุตูุงุช Edge TTS (ูููู ุชุฒูุฏู ูููู ุจุนุฏูู)
voices = {
    "ar": {
        1: "ar-EG-ShakirNeural",   # ููุฏ
        2: "ar-EG-SalmaNeural",    # ุจูุช
        3: "ar-SA-HamedNeural"     # ุตูุช ุฑุฌููู/ูููุฒ (ูุฑูุจ ูููุฑุชููู)
    },
    "en": {
        1: "en-US-GuyNeural",      # ููุฏ
        2: "en-US-JennyNeural",    # ุจูุช
        3: "en-GB-RyanNeural"      # ุตูุช ูููุฒ (ูููู ุชุญุณูู ูุฑุชููู)
    }
}

selected_voice = voices.get(lang, {}).get(voice_choice, "en-US-JennyNeural")
```

```python
# ุชุญููู ุงููุต ุฅูู ุตูุช
async def tts_edge(text, voice, rate="+0%"):
    file_path = f"audio_stories/{selected_title}.mp3"
    text_file = f"text_stories/{selected_title}.txt"

    # ุญูุธ ุงููุต
    with open(text_file, "w", encoding="utf-8") as f:
        f.write(text)

    # ุญูุธ ุงูุตูุช
    communicate = edge_tts.Communicate(text, voice=voice, rate=rate)
    await communicate.save(file_path)

    print(f"\nโ ุชู ุฅูุดุงุก ุงููุตุฉ: {selected_title} ุจุงูุตูุช: {voice}")
    display(Audio(file_path, autoplay=True))

# ุชุดุบูู ุงููุตุฉ
await tts_edge(selected_content, selected_voice)
```
